Parse through the response to get the POI (such as restaurants, bars, etc) details you want (ratings, name, location, etc)

```python
#pretty_json = json.dumps(json_data, indent=2)
#print(pretty_json)
POI=json_data
POI
```

```python
#category_name
#category_name=POI["results"][0]["categories"][0]["name"]
category_name_list=[]
for objects in POI["results"]:
    category_name=objects["categories"][0]["name"]
    category_name_list.append(category_name)
   
     
#distance=POI["results"][0]["distance"]
#distance
distance_list=[]
for objects in POI["results"]:
    distance=objects["distance"]
    distance_list.append(distance)
    
    
    
#latitude=POI["results"][0]["geocodes"]["main"]["latitude"]
#latitude
latitude_list=[]
for objects in POI["results"]:
    latitude=objects["geocodes"]["main"]["latitude"]
    latitude_list.append(latitude)
   
    
#longitude=POI["results"][0]["geocodes"]["main"]["longitude"]
#longitude
longitude_list=[]
for objects in POI["results"]:
    longitude=objects["geocodes"]["main"]["longitude"]
    longitude_list.append(longitude)
    

#locality=POI["results"][0]["location"]["locality"]
#locality
locality_list=[]
for objects in POI["results"]:
    locality=objects["location"]["locality"]
    locality_list.append(locality)
    
    
#postcode=POI["results"][0]["location"]["postcode"]
#postcode
postcode_list=[]
for objects in POI["results"]:
    postcode=objects["location"].get("postcode")
    postcode_list.append(postcode)
    

#region=POI["results"][0]["location"]["region"]
#region
region_list=[]
for objects in POI["results"]:
    region=objects["location"]["region"]
    region_list.append(region)
    
    
#place_name=POI["results"][0]["name"]
#place_name
place_name_list=[]
for objects in POI["results"]:
    place_name=objects["name"]
    place_name_list.append(place_name)
    
    
#rating=POI["results"][0]["rating"]
#rating
rating_list=[]
for objects in POI["results"]:
    rating=objects["rating"]
    rating_list.append(rating)
```

Put your parsed results into a DataFrame

```python
POI_data={
    "business_name":place_name_list,
    "category_name":category_name_list,
    "distance":distance_list,
    "rating":rating_list,
    "latitude":latitude_list,
    "longitude":longitude_list,  
    "locality":locality_list,
    "region":region_list,
    "postcode":postcode_list,        
}

# index set to 50 since we have limit of 50 business from Foursquare
index_values = []
for i in range (1,51):
    index_values.append(i)

# create FourSquare dataframe    
FourSqr_df=pd.DataFrame(POI_data,index=index_values)
FourSqr_df.head(10)
FourSqr_df.to_csv(r"C:\Users\Anna\OneDrive\Desktop\New folder\FourSqr_orig.csv", index=False)
```
