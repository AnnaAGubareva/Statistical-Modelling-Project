```python
#SQL database

# create new database 
conn = sqlite3.connect("mydatabaseSql.db")


# Create a cursor object to execute SQL commands
cursor = conn.cursor()

# Create a Bike table
cursor.execute('''
    CREATE TABLE Bike_table (
        empty_slots INTEGER,
        free_bikes INTEGER,
        total_slots INTEGER
        name TEXT,
        latitude FLOAT,
        longitude FLOAT,
        difference_in_m FLOAT
    )
''')

# Create a FourSqr table
cursor.execute('''
    CREATE TABLE FourSqr_table (
     business_name TEXT,
     category_name TEXT,
        distance FLOAT,
        rating FLOAT,
        latitude FLOAT,
        longitude FLOAT,
        locality TEXT,
        region TEXT,
        postcode INTEGER
    )
''')

# Create a Yelp table
cursor.execute('''
    CREATE TABLE Yelp_table (
        business_name TEXT,
        title TEXT,
        distance FLOAT,
        rating FLOAT,
        review_count INTEGER
        latitude FLOAT,
        longitude FLOAT,
        zipcode INTEGER
    )
''')

# Commit the changes and close the connection
conn.commit()

# Close the connection
conn.close()
```

```python
conn = sqlite3.connect("mydatabaseSql.db")

# Insert data into the database
bike_df.to_sql("Bike_table", conn, if_exists="replace", index=False)
fourSqr_df.to_sql("FourSqr_table", conn, if_exists="replace", index=False)
yelp_df.to_sql ("Yelp_table", conn, if_exists="replace", index=False)

# Commit the changes and close the connection
conn.commit()
conn.close()
```
